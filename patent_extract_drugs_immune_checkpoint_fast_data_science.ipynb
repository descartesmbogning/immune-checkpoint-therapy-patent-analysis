# extraction

```python
import pandas as pd
import re

def extract_unique_drug_names(df, text_column):
    # Define the pattern for tokenization
    PATTERN = r"\w+"

    # Function to find drugs in a text
    def find_drugs_in_text(text):
        tokens = re.findall(PATTERN, text)
        drugs_found = find_drugs(tokens)
        # Extract only the drug names, remove duplicates, and join with semicolon
        drug_names = '; '.join(list(set([drug_info[0]['name'] for drug_info in drugs_found])))
        return drug_names

    # Apply the function to the text column of the dataframe
    df['drugs_fast_data_science'] = df[text_column].apply(find_drugs_in_text)

    return df
```

```python
df_with_unique_drug_names = extract_unique_drug_names(df, 'Title_Abstract_concat')
df_with_unique_drug_names
```

```python
df_with_unique_drug_names[['drugs_fast_data_science']]
```

```python
df2 =df_with_unique_drug_names.copy()
# Split the semicolon-separated strings in the 'drugs' column into lists
df2['drugs_fast_data_science'] = df2['drugs_fast_data_science'].apply(lambda x: x.split('; '))

# Flatten the lists in the 'drugs' column into a single list
all_drugs = [drug for sublist in df2['drugs_fast_data_science'].tolist() for drug in sublist]

# Remove duplicates from the list of all drugs
unique_drugs = list(set(all_drugs))

unique_drugs
```
