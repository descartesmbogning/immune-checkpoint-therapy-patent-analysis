```python
import pandas as pd
import os
from spacy.lang.en.stop_words import STOP_WORDS
import spacy
import re

# Assuming nlp is already loaded with your NLP model
# nlp = ...
nlp = spacy.load("en_core_sci_sm")

def extract_biomedical_data(text):
    nlp = spacy.load("en_core_sci_sm")
    doc = nlp(text)
    biomed_data = doc.ents
    # Convert all elements to lowercase
    lower_tuples = [str(x).lower() for x in biomed_data]

    # Remove duplicates by converting the list to a set
    unique_tuples = set(lower_tuples)

    # Remove stop words
    relevant_expressions = [expr for expr in unique_tuples if not any(word in STOP_WORDS for word in expr.split())]

    # Keep only expressions with specific structure (alphanumeric and possibly hyphens or spaces)
    structured_expressions = [expr for expr in relevant_expressions if re.match(r'^[\w\s-]+$', expr)]

    # List of expressions to exclude from lemmatization
    exclusions = ["17-aag", "pd-1", "pd-l1", "ctla-4", "17 aag", "pd 1", "pd l1", "ctla 4", "17aag", "pd1", "pdl1", "ctla4",'Pd-1','Pd 1','Pd1']

    # Lemmatization
    nlp = spacy.load("en_core_web_sm")
    lemmatized_expressions = [' '.join([nlp(word)[0].lemma_ if expr not in exclusions else word for word in expr.split()]) for expr in structured_expressions]

    # Join the unique elements with ';'
    result = '; '.join(lemmatized_expressions)

    return result

def process_chunk(chunk_df):
    chunk_df['biomedical_data'] = chunk_df['Title_Abstract_concat'].apply(extract_biomedical_data)
    return chunk_df

# Assuming df is your DataFrame with concatenated titles and abstracts
# df = ...

# Create the directory if it doesn't exist
output_directory = "/content/drive/Othercomputers/My laptop/D:/PROJECT/YORK UNIVERSITY/Immune checkpoint /data_preprocesse/biomedical_data_files"
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Split dataframe into chunks of 500 rows
chunks = [df.iloc[i:i + 100000] for i in range(0, df.shape[0], 10000)]

for idx, chunk in enumerate(chunks):
    # Apply the function to extract taxons
    processed_chunk = process_chunk(chunk)

    # Save the chunk to a file
    output_path = os.path.join(output_directory, f"data_immune_checkpoints_topics_and_biomedical_data_ALL{idx}.csv")
    processed_chunk.to_csv(output_path, index=False)
    print(f"{idx} : Processing and saving completed!")

print("Processing and saving completed!")
```
