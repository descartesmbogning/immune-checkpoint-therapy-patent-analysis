# extraction

```python
import pandas as pd
import os

# Assuming nlp is already loaded with your NLP model
# nlp = ...

def extract_taxons(text):
    doc = nlp(text)
    Taxons = doc.ents
    return Taxons

def process_chunk(chunk_df):
    chunk_df['Taxons'] = chunk_df['Title_Abstract_concat'].apply(extract_taxons)
    return chunk_df

# Assuming df is your DataFrame with concatenated titles and abstracts
# df = ...

# Create the directory if it doesn't exist
output_directory = "/content/drive/Othercomputers/My laptop/D:/PROJECT/YORK UNIVERSITY/Immune checkpoint /data_preprocesse/taxa_files"
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Split dataframe into chunks of 500 rows
chunks = [df.iloc[i:i + 100] for i in range(0, df.shape[0], 100)]

for idx, chunk in enumerate(chunks):
    # Apply the function to extract taxons
    processed_chunk = process_chunk(chunk)

    # Save the chunk to a file
    output_path = os.path.join(output_directory, f"data_immune_checkpoints_topics_and_diseases_d_{idx}.csv")
    processed_chunk.to_csv(output_path, index=False)
    print(f"{idx} : Processing and saving completed!")

print("Processing and saving completed!")
```
